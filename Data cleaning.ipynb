```python
import pandas as pd
print(pd. __version__)
#load the data
data = pd.read_csv("thesis_data.csv")
data.head()
data.shape
```

```python
#filter out unwanted columns from the dataframe
df = data.drop(index=[0,1], columns = ['StartDate', 'EndDate', 'Status', 'IPAddress', 'Finished',
       'Duration (in seconds)', 'RecordedDate', 'ResponseId',
       'RecipientLastName', 'RecipientFirstName', 'RecipientEmail',
       'ExternalReference', 'LocationLatitude', 'LocationLongitude',
       'DistributionChannel', 'UserLanguage', 'A1.aud.time_First Click', 
       'A1.aud.time_Last Click', 'A1.aud.time_Click Count', 'A1.time_Page Submit',
       'A2.aud.time_First Click', 'A2.aud.time_Last Click', 'A2.aud.time_Click Count',
       'A2.time_Page Submit', 'A3.aud.time_First Click', 'A3.aud.time_Last Click', 
       'A3.aud.time_Click Count', 'A3.time_Page Submit', 'B1.aud.time_First Click', 
       'B1.aud.time_Last Click', 'B1.aud.time_Click Count', 'B1.time_Page Submit',
       'B2.aud.time_First Click', 'B2.aud.time_Last Click', 'B2.aud.time_Click Count',
       'B2.time_Page Submit', 'B3.aud.time_First Click', 'B3.aud.time_Last Click', 
       'B3.aud.time_Click Count', 'B3.time_Page Submit', 'C1.aud.time_First Click', 
       'C1.aud.time_Last Click', 'C1.aud.time_Click Count', 'C1.time_Page Submit',
       'C2.aud.time_First Click', 'C2.aud.time_Last Click', 'C2.aud.time_Click Count',
       'C2.time_Page Submit', 'C3.aud.time_First Click', 'C3.aud.time_Last Click', 
       'C3.aud.time_Click Count', 'C3.time_Page Submit', 'D1.aud.time_First Click', 
       'D1.aud.time_Last Click', 'D1.aud.time_Click Count', 'D1.time_Page Submit',
       'D2.aud.time_First Click', 'D2.aud.time_Last Click', 'D2.aud.time_Click Count',
       'D2.time_Page Submit', 'D3.aud.time_First Click', 'D3.aud.time_Last Click', 
       'D3.aud.time_Click Count', 'D3.time_Page Submit', 'email', 'open'])

df.head()
```

```python
#recode columns that have numeric values as the categories back to the label name
df['Gender'] = pd.to_numeric(df['Gender'])
df.Gender.replace({1:"Man", 4:"Woman", 5:"Non-binary", 6:"Other"}, inplace=True)

df['Age'] = pd.to_numeric(df['Age'])
df.Age.replace({1:"below 18", 2:"18-23", 3:"24-30", 4:"31-36", 5:"37-42", 6:"43-49", 7:"50+"}, inplace=True)

df['EngProf'] = pd.to_numeric(df['EngProf'])
df.EngProf.replace({1:"Native/Bilingual", 2:"Professional Working Proficiency", 
                    3:"Limited working Proficiency",4:"Elementary Proficiency"}, inplace=True)
df['EngFL'] = pd.to_numeric(df['EngFL'])
df.EngFL.replace({1:"Yes", 2:"No"}, inplace=True)
```

```python
#change data type of columns 'Progress' and 'Consent' to numeric.
df['Progress'] = pd.to_numeric(df['Progress'])
df['Consent'] = pd.to_numeric(df['Consent'])

#change data type of all columns containing the amount of time the participants spent listened to the audio to numeric.
df['A1.aud.time_Page Submit'] = pd.to_numeric(df['A1.aud.time_Page Submit'])
df['A2.aud.time_Page Submit'] = pd.to_numeric(df['A2.aud.time_Page Submit'])
df['A3.aud.time_Page Submit'] = pd.to_numeric(df['A3.aud.time_Page Submit'])
df['B1.aud.time_Page Submit'] = pd.to_numeric(df['B1.aud.time_Page Submit'])
df['B2.aud.time_Page Submit'] = pd.to_numeric(df['B2.aud.time_Page Submit'])
df['B3.aud.time_Page Submit'] = pd.to_numeric(df['B3.aud.time_Page Submit'])
df['C1.aud.time_Page Submit'] = pd.to_numeric(df['C1.aud.time_Page Submit'])
df['C2.aud.time_Page Submit'] = pd.to_numeric(df['C2.aud.time_Page Submit'])
df['C3.aud.time_Page Submit'] = pd.to_numeric(df['C3.aud.time_Page Submit'])
df['D1.aud.time_Page Submit'] = pd.to_numeric(df['D1.aud.time_Page Submit'])
df['D2.aud.time_Page Submit'] = pd.to_numeric(df['D2.aud.time_Page Submit'])
df['D3.aud.time_Page Submit'] = pd.to_numeric(df['D3.aud.time_Page Submit'])

#Filter columns 'Progress' and 'Consent' to only include data of those who consented to and completed at least 2 out of 3 of the audio tasks.
df = df[(df['Progress'] > 75) & (df['Consent'] == 1)]

#Filter out participants that did not listen to the full audio
df = df[((df['A1.aud.time_Page Submit'] > 164) | (df['A2.aud.time_Page Submit'] > 222) | (df['A3.aud.time_Page Submit'] > 195)) |
                  ((df['B1.aud.time_Page Submit'] > 216) | (df['B2.aud.time_Page Submit'] > 186) | (df['B3.aud.time_Page Submit'] > 203)) |
                  ((df['C1.aud.time_Page Submit'] > 184) | (df['C2.aud.time_Page Submit'] > 196) | (df['C3.aud.time_Page Submit'] > 168)) |
                  ((df['D1.aud.time_Page Submit'] > 191) | (df['D2.aud.time_Page Submit'] > 164) | (df['D3.aud.time_Page Submit'] > 218))]
```

```python
'''The rest of the code is going to clean the data. We will restructure the data to have a column named "Version"
which will contain the version of the experiment that each participant participated in. By doing so we can get rid 
of the multiple cells in each row that contain NaN values that correspond to the other versions of the experiment.'''

#create 4 subsets of the data for each version of the survey
x = df.loc[:,'Gender': 'SC2'].copy()  #all the demographic data
a = df.loc[:,'A1.aud.time_Page Submit':'A3.att_4'].copy()  #all the columns for the specified version
b = df.loc[:,'B1.aud.time_Page Submit':'B3.att_4'].copy()
c = df.loc[:,'C1.aud.time_Page Submit':'C3.att_4'].copy()
d = df.loc[:,'D1.aud.time_Page Submit':'D3.att_4'].copy()

df_A = pd.concat([a,x],axis=1)   #merge the columns of the version specific data and demographic data 
df_B = pd.concat([b,x],axis=1)
df_C = pd.concat([c,x],axis=1)
df_D = pd.concat([d,x],axis=1)

#add a new column "Version" that has the label of the version of each subset.
df_A["Version"] = ['A'] * df_A.shape[0]
df_B["Version"] = ['B'] * df_B.shape[0]
df_C["Version"] = ['C'] * df_C.shape[0]
df_D["Version"] = ['D'] * df_D.shape[0]
```

```python
subsets = [df_A, df_B, df_C, df_D]

#loop through all 4 subsets deleting the rows containing NaN values for that version
#The discarded rows are of participants that did not fill in that version of the experiment
for s in subsets:
    for index, row in s.iterrows():
        if pd.isnull(row[1]) & pd.isnull(row[14]) & pd.isnull(row[27]):
            s.drop(index, inplace=True)
```

```python
#create a new list of column names
cols = ['Aud_time1', '1_1',	'1_2', '1_3', 'fc_1', 'lc_1', 'cc_1', 'Pleasant_1',	'Natural_1', 'Human_1',	'Comp_1', 'Cred_1',	'Attention_Human',	
'Aud_time2', '2_1',	'2_2', '2_3', 'fc_2', 'lc_2', 'cc_2', 'Pleasant_2',	'Natural_2', 'Human_2',	'Comp_2', 'Cred_2',	'Attention_AI',	
'Aud_time3', '3_1',	'3_2', '3_3', 'fc_3', 'lc_3', 'cc_3', 'Pleasant_3',	'Natural_3', 'Human_3',	'Comp_3', 'Cred_3',	'Attention_TTS',
'Gender', 'Age', 'EngFL', 'EngProf', 'SC_Human', 'SC_AI', 'SC_TTS', 'Version']

#rename the column names for all 4 subsets (helps concatenate them by matching the column names)
df_A.columns = cols
df_B.columns = cols
df_C.columns = cols
df_D.columns = cols
```

```python
#merge all 4 subsets into 1 new dataframe
df_clean = pd.concat([df_A, df_B, df_C, df_D])

#convert the index into a column
df_clean.reset_index(inplace = True)

#reset the values of the index column to 1,2,3,4,...
df_clean['index'] = range(1,df_clean.shape[0]+1)

#change the column name to 'Participant_ID'
df_clean.rename(columns={'index': 'Participant_ID'}, inplace=True)

df_clean
```

```python
#download the cleaned data as a .csv file
df_clean.to_csv("Thesis_data_clean.csv")
```
