```python
import matplotlib.pyplot as plt
import pandas as pd
from statistics import mean

data = pd.read_csv("Thesis_data_clean.csv")
```

```python
df = data.loc[:,["Attention_Human", "Attention_AI", "Attention_TTS"]]
df.rename(columns={'Attention_Human': 'Human', 'Attention_AI':'AI', 'Attention_TTS':'TTS'}, inplace=True)

ax = df.mean().plot(y=["Human", "AI", "TTS"],
        kind="bar", figsize=(10, 5), xlabel="Condition", ylabel="Attention rating (out of 100)")
ax.bar_label(ax.containers[0])
```

```python
df = data.loc[:,"SC_Human":"SC_TTS"]
df.rename(columns={'SC_human': 'Human', 'SC_AI':'AI', 'SC_TTS':'TTS'}, inplace=True)

ax = df.mean().plot(y=["Human", "AI", "TTS"], kind="bar", figsize=(10, 5), 
               xlabel="Condition", ylabel="Average Score", yticks=[1.5,2,2.5])
ax.bar_label(ax.containers[0])
```

```python
df_comp = data.loc[:,['Comp_1', 'Comp_2', 'Comp_3']]
df_comp.rename(columns={'Comp_1': 'Human', 'Comp_2':'AI', 'Comp_3':'TTS'}, inplace=True)
df_nat = data.loc[:,['Natural_1', 'Natural_2', 'Natural_3']]
df_nat.rename(columns={'Natural_1': 'Human', 'Natural_2':'AI', 'Natural_3':'TTS'}, inplace=True)
df_hum = data.loc[:,['Human_1', 'Human_2', 'Human_3']]
df_hum.rename(columns={'Human_1': 'Human', 'Human_2':'AI', 'Human_3':'TTS'}, inplace=True)
df_ples = data.loc[:,['Pleasant_1', 'Pleasant_2', 'Pleasant_3']]
df_ples.rename(columns={'Pleasant_1': 'Human', 'Pleasant_2':'AI', 'Pleasant_3':'TTS'}, inplace=True)
df_cred = data.loc[:,['Cred_1', 'Cred_2', 'Cred_3']]
df_cred.rename(columns={'Cred_1': 'Human', 'Cred_2':'AI', 'Cred_3':'TTS'}, inplace=True)

              
df_comp.mean().plot(y=["Human", "AI", "TTS"], kind="line", xlabel="Condition", ylabel="Rating")
df_nat.mean().plot(y=["Human", "AI", "TTS"], kind="line", xlabel="Condition")
df_hum.mean().plot(y=["Human", "AI", "TTS"], kind="line", xlabel="Condition")
df_ples.mean().plot(y=["Human", "AI", "TTS"], kind="line", xlabel="Condition")
df_cred.mean().plot(y=["Human", "AI", "TTS"], kind="line", xlabel="Condition")

plt.legend(['Comprehension', 'Naturalness', 'Human-likeness', 'Pleasantness', 'Credability'])
plt.savefig("Ratings.png")
```
